```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        countries.append(country)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
city_data = list(zip(cities, countries))
city_data_df = pd.DataFrame(city_data, columns = ['City', 'Country'])
city_data_df
```

```python
# get actual lat and lng for each city found using Weather API
city_data_df['Latitude'] = ""
city_data_df['Longitude'] = ""
base_url = 'http://api.openweathermap.org/data/2.5/weather?q'
units = 'imperial'
city_data_df
```

```python
for index1, rows in city_data_df.iterrows():
    lat_lng_url = f'{base_url}={rows[0]}&appid={weather_api_key}&units={units}'
    response1 = requests.get(lat_lng_url).json()

    try:
        city_data_df.loc[index1, 'Latitude'] = response1['coord']['lat']
        city_data_df.loc[index1, 'Longitude'] = response1['coord']['lon']
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    time.sleep(1)
```

```python
city_data_df
```

```python
# Remove rows with no values
# Set empty strings to 'NaN'
nan_value = float("NaN")
city_data_df.replace("", nan_value, inplace = True)
# Drop rows with any NaN
city_data_df.dropna(how = 'any', inplace = True)
city_data_df
```

```python
# Set base url for OpenCall API (needs lat and lng to search)
url = 'http://api.openweathermap.org/data/2.5/onecall?'
units = 'imperial'

# Add new values to main dataframe
city_data_df['Date'] = ""
city_data_df['Temp (Day)'] = ""
city_data_df['Humidity'] = ""
city_data_df['Cloudiness'] = ""
city_data_df['Wind Speed'] = ""
city_data_df.head()
```

```python
# Search for weather info using OpenCall API
# I wanted to get forecast data for the next day so that all data points would be drawn from the same time of day.
# If I used current temps, some locations would be at night and some in the middle of the day.

for index, row in city_data_df.iterrows():
    query_url = f'{url}appid={weather_api_key}&lat={row[2]}&lon={row[3]}&units={units}'
    print(f"Retrieving Results for Index {index}: {row[0]}.")
    response = requests.get(query_url).json()
    

    try:
        print("All data found.")
        
        city_data_df.loc[index, 'Date'] = response['daily'][0]['dt']
        city_data_df.loc[index, 'Temp (Day)'] = response['daily'][0]['temp']['day']
        city_data_df.loc[index, 'Humidity'] = response['daily'][0]['humidity']
        city_data_df.loc[index, 'Cloudiness'] = response['daily'][0]['clouds']
        city_data_df.loc[index, 'Wind Speed'] = response['daily'][0]['wind_speed']
        
    except (KeyError, IndexError):
        print("City not found... skipping.")
    
    time.sleep(1)    
    print("------------")
```

```python
# Create Northern and Southern Hemisphere DataFrames
city_north = city_data_renamed.loc[city_data_renamed['Latitude'] >= 0]
display(city_north.describe())
city_south = city_data_renamed.loc[city_data_renamed['Latitude'] < 0]
display(city_south.describe())
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
xnt_values = city_north['Latitude']
ynt_values = city_north['Temp (Day) (F)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(xnt_values, ynt_values)
regress_values = xnt_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(xnt_values,ynt_values)
plt.plot(xnt_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title('Northern Hemisphere City Latitude vs. Temperature in F (04/14/2020)')
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.savefig("./Images/NorthLatitudeVsTemp.png")
plt.show()
print(f"The r-squared is: {rvalue}.")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
xst_values = city_south['Latitude']
yst_values = city_south['Temp (Day) (F)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(xst_values, yst_values)
regress_values = xst_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(xst_values,yst_values)
plt.plot(xst_values,regress_values,"r-")
plt.annotate(line_eq,(-25, 50),fontsize=15,color="red")
plt.title('Southern Hemisphere City Latitude vs. Temperature in F (04/14/2020)')
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.savefig("./Images/SouthLatitudeVsTemp.png")
plt.show()
print(f"The r-squared is: {rvalue}.")
```

The temperatures in the northern hemisphere appear to be positively correlated to the latitude, given that the r-squared value is 88%. The southern himisphere values have a slightly lower r-squared (67%), but they still show positive correlation to the latitudes.
